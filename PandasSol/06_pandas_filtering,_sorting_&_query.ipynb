```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/tukkaLearn/datasets/refs/heads/main/ecommerce_full_2025_generated.csv")
df.head()
```

### Simple filter – high value orders only


```python
mask=df["price"]>1000
df[mask][["order_id","customer_id","product_name","price"]]
```

### Multiple conditions with & and |


```python
mask=((df["country"] =="USA") | (df["country"] =="India")) & (df["price"]>300) & (df["category"]=="Electronics")
display(df[mask])

# Or Use Query

df_filter=df.query("(country == 'India' | country == 'USA') & price >300 &  category == 'Electronics'")
df_filter
```

### Filter using .isin() – multiple values


```python
df[df["customer_id"].isin(["CUST9600",'CUST3441'])]
```

### String filtering with str.contains()


```python
df[df["product_name"].str.contains("iPhone|MacBook",case=False,na=False)]
```

### Date range filtering


```python
# Change To DateTime
df["order_date"]=pd.to_datetime(df["order_date"])

df_filtered = df[(df["order_date"] >= "2025-11-28") & (df["order_date"] <= "2025-12-02")]
display(df_filtered)

# Or Use
df[df["order_date"].between("2025-11-28","2025-12-01")]
```

### Sort Data


```python
df["total_amount"]=df["price"]*df["quantity"]

filterWithPrice=df.sort_values(by=["total_amount","customer_id"],ascending=False)
filterWithPrice[["order_id","product_name","total_amount"]]
```

### Powerful .query()


```python
countryList=["USA","Canada"]
status=["Delivered","Return"]

df_filter=df.query("(country == @countryList[0] | country == @countryList[1]) & (price >500) & (status == @status[0] | status == @status[1]) &  discount < 0.2")
df_filter
```

### Filter + sort + head


```python
df_vip_customer=df[df["total_amount"]>3000]
df_vip_customer=df_vip_customer.sort_values(by="total_amount",ascending=False)
df_vip_customer[["customer_id","quantity","total_amount"]]

# Single Line

# df[df["total_amount"]>3000].sort_values(by="total_amount",ascending=False)[["customer_id","quantity","total_amount"]]
```

### Complex query() with variables


```python
min_amount = 1000
countries = ['USA', 'India', 'Germany']

df_filter= df.query("status == 'Delivered' & price < @min_amount & (country == @countries[0] | country == @countries[1] | country == @countries[2])")

df_filter.sort_values(by='total_amount',ascending=False)
```

### filter + query + sort


```python
df_pending_order=df.query("status == 'Pending' & total_amount > 2000 ").sort_values(by="order_date")
df_pending_order[["order_id","order_date","status",'total_amount']]
```
