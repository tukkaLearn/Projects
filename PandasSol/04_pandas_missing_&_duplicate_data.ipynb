```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/tukkaLearn/datasets/refs/heads/main/orders_with_problems_nov2025.csv')
df.head()
```

### How bad is the missing data?


```python
print("Most missing values ")
display(df.isnull().sum().sort_values(ascending=False).head(3))

print("Percent missing values ")

display(df.isnull().sum().sort_values(ascending=False).head(3)/df.shape[0]*100)
```

### Keep only complete customer records


```python
print("Total Customer Before Clean ")

display(len(df))


df2 = df.dropna(subset=['email', 'country'])
print("Total Customer After Clean ")

display(len(df2))
```

```python
df
```

### Fill missing prices intelligently


```python
df[df["price"].isnull()]
```

```python
med=df[df["product_category"]=="Fashion"]["price"].median()
df["price"].fillna(med)
df
```

### Forward-fill missing status in time-series


```python
df["status"].ffill(inplace=True)
df
```

### Backward-fill delivery dates


```python
df["delivery_date"].bfill(inplace=True)
df
```

### Find duplicate orders (exact match)


```python
df[df.duplicated(subset=["customer_id","order_date"],keep=False)]
```

```python
df.drop_duplicates(subset=["customer_id","order_date"],keep="last", inplace=True)
df
```

### Replace messy text values


```python
df["country"].replace({
    "United States": "USA",
    "USA": "USA",
    "US": "USA",
    "U.S.": "USA"
},inplace=True)
df
```

### Use where()


```python
df["price"]=df["price"].where(df["price"]>0,99.99)
df
```
