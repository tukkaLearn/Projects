### Create a beautiful data quality report


```python
report = pd.DataFrame({
    'column_name': df.columns,
    'data_type': df.dtypes.values,
    'missing_values': df.isna().sum().values,
    'percentage_missing': (df.isna().mean() * 100).values,
    'unique_values': df.nunique().values
})
```

```python
report
```
