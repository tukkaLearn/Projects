```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/tukkaLearn/datasets/refs/heads/main/ecommerce_nov2025_messy.csv')
df.head()
```

### Change one value using .at[]


```python
print("Before:")
display(df[df['customer_id'] == 'CUST_909'])

idx = df[df['customer_id'] == 'CUST_909'].index[0]
df.at[idx, 'product_name'] = 'Unknown Product'
df.at[idx, 'status'] = 'Cancelled - Invalid'

print("\nAfter fixing:")
display(df.loc[idx:idx])
```

### Replace all missing emails


```python
print(f"Missing emails before: {df['email'].isna().sum()}")

df['email'] = df['email'].fillna('no_email@provided.com')

print(f"Missing emails after: {df['email'].isna().sum()}")
df.head(8)[['customer_id', 'email']]
```

### Fix inconsistent category names


```python
print("Before:")
print(df['category'].value_counts().head())

df['category'] = df['category'].replace({
    'electronics': 'Electronics',
    'ELECTRONICS': 'Electronics',
    'Fashion ': 'Fashion',
    'fashion': 'Fashion'
})

print("\nAfter cleaning:")
print(df['category'].value_counts())
```

### Clean the price column – remove $ and commas


```python
print("Before dtype:", df['price'].dtype)
print(df['price'].head(10).tolist())

df['price'] = pd.to_numeric(
    df['price'].astype(str).str.replace(r'[^0-9.]', '', regex=True),
    errors='coerce'
)

print("\nAfter cleaning:")
print("dtype:", df['price'].dtype)
print(df['price'].describe().round(2))
```

### Add total_amount column


```python
df['total_amount'] = df['price'] * df['quantity']

print("Top 5 most valuable orders:")
display(df[['order_id', 'product_name', 'price', 'quantity', 'total_amount']]
        .sort_values('total_amount', ascending=False).head(5))
```

### Drop garbage rows and columns


```python
before = df.shape
print(f"Before: {before}")

# Drop fake zero-quantity orders
df = df[df['quantity'] > 0].copy()

# Drop useless column
df = df.drop(columns=['signup_date'], errors='ignore')

after = df.shape
print(f"After: {after}")
print(f"Removed {before[0] - after[0]} garbage rows")
```

### Rename columns to snake_case


```python
df = df.rename(columns={
    'product_name': 'product',
    'payment_method': 'payment',
    'order_date': 'date'
})

print("Clean professional column names:")
print(df.columns.tolist())
```

### Type casting party!


```python
df['price'] = df['price'].astype('float32')
df['quantity'] = df['quantity'].astype('int32')
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['category'] = df['category'].astype('category')
df['status'] = df['status'].astype('category')

print("Memory dropped from", round(df.memory_usage(deep=True).sum() / 1024**2, 2), "MB → optimized!")
df.info()
```

### Replace weird payment methods


```python
print("Before:")
print(df['payment'].value_counts())

df['payment'] = df['payment'].replace({
    'credit card': 'Credit Card',
    'CREDIT': 'Credit Card',
    'paypal': 'PayPal',
    'CASH': 'Cash'
})

print("\nAfter:")
print(df['payment'].value_counts())
```
