```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/tukkaLearn/datasets/refs/heads/main/orders_full_nov2025.csv")
df.head()
```

### Simple map() – turn category name into codes


```python
codes={'Electronics': "E",
 'Fashion': "F",
 'Gaming': "G",}
```

```python
df["category_codes"]=df["category"].map(codes)
```

```python
df["category_codes"].value_counts()
```

### map() with function – extract email domain


```python
def extractEmailDomain(email):
    if pd.isnull(email):
        return None
    return email.split("@")[-1]

df["email_domain"]=df["email"].map(lambda email: extractEmailDomain(email))
```

```python
df["email_domain"].value_counts()
```

### apply() on single column – calculate shipping delay


```python
# Change To DateTime
df["order_date"]=pd.to_datetime(df["order_date"])

df["delivery_date"]=pd.to_datetime(df["delivery_date"])

df["delay_days"]=df["delivery_date"] - df["order_date"]
df["delay_days"]=df["delay_days"].apply(lambda x: x.days)

#  OR use 
# df["delay_days"]=(df["delivery_date"]-df["order_date"]).dt.days
df[['delay_days']]
```

### Binning – turn continuous into categories


```python
bins = [-np.inf, 500, 2000, 5000, np.inf]
labels = ['Bronze', 'Silver', 'Gold', 'Platinum']
df["customer_segments"]=pd.cut(df["total_amount"],bins=bins,labels=labels,right=False)
df["customer_segments"].value_counts()
```

### qcut() – equal-sized bins (quantile binning)


```python
df["customer_spend"]=pd.qcut(df["total_amount"],q=4,labels=["Low","Medium","High","Very High"])

df[df["customer_spend"]=="Medium"]
```
