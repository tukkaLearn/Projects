```python
import numpy as np
```

## 1. Convert sensor voltage to power


```python
voltageArray = np.array([1.5, 2.0, 3.3])
resistance = 10.0  
powerWatts = np.power(voltageArray, 2) / resistance
np.set_printoptions(precision=2)
print("Power (W):", powerWatts)
```

## 2. Apply log scaling to audio signal


```python
audioSamples = np.array([100, 1000, 10000])
dbScale = np.log10(audioSamples)
print("dB-like scale:", dbScale)
```

## 3. Normalize exponential growth


```python
bacteriaCounts = np.array([2, 8, 32, 128])
logGrowth = np.log(bacteriaCounts)
print("Log-linearized:", logGrowth)
```

## 4. Compute distance using Pythagoras


```python
deltaX = np.array([3, 4])
deltaY = np.array([4, 3])
distance = np.sqrt(np.power(deltaX, 2) + np.power(deltaY, 2))
print("Distances:", distance)
```

## 5. Generate sine wave for signal


```python
timeVector = np.linspace(0, 1, 5)
amplitude = 2.0
frequency = 1.0
sineWave = amplitude * np.sin(2 * np.pi * frequency * timeVector)
print("Sine wave:", sineWave)
```

## 6. Convert GPS heading from degrees


```python
headingsDeg = np.array([0, 90, 180, 270])
headingsRad = np.deg2rad(headingsDeg)
xComponent = np.cos(headingsRad)
print("X-component (unit vector):", xComponent)
```

## 7. Round temperature to nearest 0.5°C


```python
temperatures = np.array([23.1, 24.7, 25.3])
roundedToHalf = np.round(temperatures * 2) / 2
print("Rounded to 0.5°C:", roundedToHalf)
```

## 8. Clean negative sensor drift


```python
rawReadings = np.array([-0.1, 1.2, -0.05, 2.3])
cleanedReadings = np.abs(rawReadings)
print("Drift-corrected:", cleanedReadings)
```

## 9. Floor timestamp to minute


```python
unixTimestamps = np.array([1617203595.7, 1617203610.2])
minuteFloor = np.floor(unixTimestamps / 60) * 60
print("Floored to minute:", minuteFloor)
```

## 10. Check if any motor current is zero


```python
motorCurrents = np.array([0, 150, 200, 0])
stalledMotor = np.any(motorCurrents == 0)
print("Stalled motor detected:", stalledMotor)
```

## 11. Validate all pressure readings


```python
pressurePsi = np.array([30, 32, 31, 29])
allValid = np.all((pressurePsi >= 28) & (pressurePsi <= 35))
print("All pressures in 28–35 psi:", allValid)
```

## 12. Filter valid pH readings


```python
phValues = np.array([6.8, 7.2, 14.5, 7.0])
validPhMask = (phValues > 0) & (phValues < 14)
validPh = phValues[validPhMask]
print("Valid pH (0–14):", validPh)
```

## 13. Challenge: Signal-to-noise ratio (SNR)


```python
signal = np.array([10, 12, 11])
noise = np.array([1, 1.5, 1.2])
snrDb = 20 * np.log10(signal / noise)
print("SNR (dB):", snrDb)
```

## 14. Challenge: Safe log on zero/negative


```python
values = np.array([0, 1, 100])
safeLog = np.log1p(values)  # log(1 + x)
print("Safe log(1+x):", safeLog)
```

## 15. Challenge: Angle normalization


```python
rawAngles = np.array([370, -90, 450])
normalizedAngles = np.mod(rawAngles, 360)
print("Normalized [0,360):", normalizedAngles)
```
