## 1. Access latest temperature reading


```python
import numpy as np
import random
temps = np.array([22.5, 23.1, 24.0, 25.2, 26.8])
latest = temps[-1]
print("Latest reading:", latest)
```

## 2. Extract pixel from satellite image


```python
import random

img= [[random.randint(0, 255) for _ in range(512)] for _ in range(512)]

img = np.array(img, dtype=np.uint8)
pixel = img[100, 200]
print("Pixel at (100, 200):", pixel)
```

## 3. Get middle row of ECG signal


```python
ecg = [[random.uniform(-1, 1) for _ in range(1000)] for _ in range(8)]
ecg = np.array(ecg)
channel4 = ecg[3]
print("First 5 values of channel 4:", channel4[:5])
```

## 4. Crop face region from image


```python
rgb_img = [[[np.random.randint(0, 256) for _ in range(3)]for _ in range(600)] for _ in range(400)]

rgb_img = np.array(rgb_img, dtype=np.uint8)

face_crop = rgb_img[150:250, 250:350]
print("Cropped face shape:", face_crop)
print("Cropped face shape:", face_crop.shape)
```

## 5. Reverse video frames


```python
frames = [i for i in range(300)]
frames = np.array(frames)
reversed_frames = frames[::-1]
print("First 5 reversed indices:", reversed_frames[:5])
```

## 6. Extract every 10th LiDAR point


```python
lidar = [random.uniform(0,1) for _  in range(1000000)]
lidar=np.array(lidar)
downsampled = lidar[::10]
print("Downsampled length:", len(downsampled))
```

## 7. Get top-left 3×3 block from heatmap


```python
heatmap = [[random.uniform(20, 40) for _ in range(10)] for _ in range(10)]
heatmap = np.array(heatmap)
top_left_3x3 = heatmap[:3, :3]
print("Top-left 3×3:\n", top_left_3x3)
```

## 8. Select all border pixels of an image


```python
img = [[random.randint(1,100) for _ in range(200)] for _ in range(200)]
img = np.array(img)

topRow=img[0, :]   
bottomRow=img[-1, :]  
leftCol=img[1:-1, 0]  
rightCol=img[1:-1, -1]
border = np.array(list(topRow)+list(bottomRow)+list(leftCol)+list(rightCol))
print("Total border pixels:", len(border))
```

## 9. Iterate over patient vital signs


```python
vitals = np.array([
    [80, 120, 80],
    [75, 115, 78],
    [90, 130, 85],
    [68, 110, 70],
    [82, 125, 82]
])

for i, row in enumerate(vitals, 1):
    print(f"Patient {i}: {list(row)}")
```

## 10. Flatten and print all pixel intensities


```python
gray = [[random.randint(0, 255) for _ in range(4)] for _ in range(4)]
gray = np.array(gray, dtype=np.uint8)
for pixel in gray.flat:
    print(pixel, end=" ")
```

## 11. Use np.nditer() on 3D MRI volume


```python
mri = [[[random.randint(0, 1500) for _ in range(50)] for _ in range(100)] for _ in range(100)]
mri=np.array(mri,dtype=np.uint16)
count = 0
for voxel in np.nditer(mri):
    if voxel > 1000:
        count += 1
print("Voxels > 1000:", count)
```

## 12. Extract diagonal of calibration matrix


```python
calib = np.array([
    [1000, 0, 320, 0],
    [0, 1000, 240, 0],
    [0, 0, 1, 0],
    [0, 0, 0, 1]
])
diagonal = [calib[0,0],calib[1,1],calib[2,2],calib[3,3]]
print("Main diagonal:", diagonal)
```

## 13. Safe slicing with out-of-bounds


```python
arr = np.arange(10)
print("arr[20:25] ", arr[20:25]) 
print("arr[-15:-10] ", arr[-15:-10]) 
```

## 14. Memory-efficient ROI without copying


```python
img = [[random.randint(0, 255) for _ in range(3000)] for _ in range(3000)] 
img= np.array(img)
roi = img[1000:2000, 1000:2000]  
print("Is ROI a view?", roi.base is img)
print("Memory efficient! No copy unless modified.")
```

## 15. Challenge: Extract checkerboard pattern


```python
grid = [[random.randint(0, 100) for _ in range(8)] for _ in range(8)]
grid = np.array(grid)
checkerboard = grid[::2, ::2] 
print("Checkerboard (even rows/cols):\n", checkerboard)
print("Shape:", checkerboard.shape)
```
