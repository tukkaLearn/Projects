```python
import numpy as np
```

## 1. Find peak heart rate


```python
heartRatesBpm = np.array([72, 95, 110, 88, 120, 65])
peakRate = np.max(heartRatesBpm)
peakIndex = np.argmax(heartRatesBpm)
print(f"Peak heart rate: {peakRate} BPM at index {peakIndex}")
```

## 2. Total rainfall in a week


```python
dailyRainfallMm = np.array([0, 5.2, 12.1, 0, 8.7, 3.3, 1.5])
weeklyTotal = np.sum(dailyRainfallMm)
print(f"Weekly rainfall: {weeklyTotal} mm")
```

## 3. City-wise average temperature


```python
cityTemps = np.random.randint(25, 35, size=(3, 7))
cityAverages = np.mean(cityTemps, axis=1)
print("City temperatures (3×7):\n", cityTemps)
print("City averages:", cityAverages)
```

## 4. Detect sensor failure


```python
sensorReadings = np.array([100, 101, 102, 0, 99, 101])
hasZero = np.any(sensorReadings == 0)
print("Sensor failure (zero reading):", hasZero)
```

## 5. Cumulative sales growth


```python
dailySales = np.array([200, 300, 250, 400])
cumulativeSales = np.cumsum(dailySales)
print("Cumulative sales:", cumulativeSales)
```

## 6. Weighted exam score


```python
examScores = np.array([80, 90, 70])
examWeights = np.array([0.2, 0.5, 0.3])
finalGrade = np.average(examScores, weights=examWeights)
print(f"Final grade: {finalGrade:.2f}")
```

## 7. Median house price


```python
housePricesLakh = np.array([45, 60, 55, 80, 50])
medianPrice = np.median(housePricesLakh)
print(f"Median house price: ₹{medianPrice} lakh")
```

## 8. Volatility of stock returns


```python
dailyReturns = np.array([0.01, -0.02, 0.03, 0.01])
volatility = np.std(dailyReturns)
print(f"Volatility (std dev): {volatility:.4f}")
```

## 9. Find 90th percentile income


```python
incomesK = np.array([30, 45, 50, 80, 120])
top10PercentCutoff = np.percentile(incomesK, 90)
print(f"90th percentile income: ₹{top10PercentCutoff}k")
```

## 10. Daily max temperature trend


```python
maxTemps = np.array([32, 33, 35, 34, 31, 30, 32])
runningSum = np.cumsum(maxTemps)
days = np.arange(1, len(maxTemps) + 1)
runningAverage = runningSum / days
print("Running average temps:", runningAverage)
```

## 11. Correlation: ad spend vs sales


```python
adSpend = np.array([10, 20, 15, 30])
sales = np.array([50, 90, 70, 120])
correlationMatrix = np.corrcoef(adSpend, sales)
print("Correlation matrix:\n", correlationMatrix)
print(f"Ad-Sales correlation: {correlationMatrix[0,1]:.3f}")
```

## 12. Covariance of height and weight


```python
heightCm = np.array([160, 170, 165])
weightKg = np.array([55, 70, 60])
covarianceMatrix = np.cov(heightCm, weightKg)
print("Covariance matrix:\n", covarianceMatrix)
```

## 13. Column-wise product of sensor gains


```python
sensorGains = np.random.rand(3, 4)
channelProducts = np.prod(sensorGains, axis=0)
print("Sensor gains (3×4):\n", sensorGains)
print("Per-channel gain product:", channelProducts)
```

## 14. Challenge: Outlier detection using IQR


```python
data = np.array([10, 12, 15, 18, 100])
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lowerBound = q1 - 1.5 * iqr
upperBound = q3 + 1.5 * iqr
outliers = data[(data < lowerBound) | (data > upperBound)]
print(f"Q1: {q1}, Q3: {q3}, IQR: {iqr}")
print(f"Bounds: [{lowerBound}, {upperBound}]")
print("Outliers:", outliers)
```

## 15. Challenge: Moving average filter


```python
signal = np.array([1, 2, 3, 4, 5, 6])
window = 3
cumsum = np.cumsum(np.insert(signal, 0, 0))
movingAverage = (cumsum[window:] - cumsum[:-window]) / window
print("Original signal:", signal)
print(f"{window}-point moving average:", movingAverage)
```
