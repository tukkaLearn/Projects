```python
import numpy as np
```

## 1. Extract corner pixels from satellite image


```python
satelliteImage = np.random.randint(0, 256, size=(4, 4), dtype=np.uint8)
cornerRows = [0, 0, 3, 3]
cornerCols = [0, 3, 0, 3]
cornerPixels = satelliteImage[cornerRows, cornerCols]
print("4×4 Image:\n", satelliteImage)
print("Corner pixels (0,0), (0,3), (3,0), (3,3):", cornerPixels)
```

## 2. Mask cloudy regions in NDVI


```python
ndviImage = np.random.rand(100, 100)
cloudMask = ndviImage < 0.1
cloudCount = cloudMask.sum()
ndviImage[cloudMask] = 0.0  # Set clouds to 0
print(f"Cloudy pixels (NDVI < 0.1): {cloudCount}")
print("First 5×5 of cleaned NDVI:\n", ndviImage[:5, :5])
```

## 3. Select every 3rd sample from ECG


```python
ecgSignal = np.random.rand(1000)
everyThirdIndex = np.arange(0, 1000, 3)
downsampledEcg = ecgSignal[everyThirdIndex]
print(f"Original length: 1000  Downsampled length: {len(downsampledEcg)}")
```

## 4. Extract diagonal using np.ix\_


```python
matrix5x5 = np.random.randint(1, 10, size=(5, 5))
rows = [0, 2, 4]
cols = [0, 2, 4]
subBlock = matrix5x5[np.ix_(rows, cols)]
print("5×5 Matrix:\n", matrix5x5)
print("3×3 sub-block:\n", subBlock)
```

## 5. Cap sensor values > 1000


```python
sensorReadings = np.array([998, 1024, 1010, 995])
highMask = sensorReadings > 1000
sensorReadings[highMask] = 1000
print("Capped readings:", sensorReadings)
```

## 6. Pick specific cities from weather grid


```python
weatherGrid = np.random.rand(5, 24)  # 5 cities × 24 hours
cityIndices = [2, 4] 
selectedCities = weatherGrid[cityIndices, :]
print(f"Selected cities shape: {selectedCities.shape}")
```

## 7. Extract red channel from ROI


```python
rgbImage = np.random.rand(100, 100, 3)
roiMask = np.random.choice([False, True], size=(100, 100), p=[0.7, 0.3])
redChannelRoi = rgbImage[roiMask, 0]
print(f"Red values in ROI: {len(redChannelRoi)} pixels")
```

## 8. Select non-contiguous frames from video


```python
videoFrames = np.random.rand(300, 64, 64)  # 300 frames
keyFrameIndices = [10, 50, 100, 200, 290]
keyFrames = videoFrames[keyFrameIndices]
print("Key frame indices:", keyFrameIndices)
print("Key frames shape:", keyFrames.shape)
```

## 9. Set invalid GPS to NaN


```python
gpsData = np.random.rand(100, 2) * 100
gpsData[::10] = [0, 0]  # Simulate invalid points
invalidMask = (gpsData[:, 0] == 0) | (gpsData[:, 1] == 0)
invalidCount = invalidMask.sum()
gpsData[invalidMask] = np.nan
print(f"Invalid GPS points: {invalidCount}")
```

## 10. Mixed indexing: top-left 2×2, skip even columns


```python
grid6x6 = np.random.randint(1, 100, size=(6, 6))
selectedRows = slice(0, 2)
selectedCols = [1, 3, 5]
subGrid = grid6x6[selectedRows, selectedCols]
print("6×6 Grid:\n", grid6x6)
print("2×3 sub-grid:\n", subGrid)
```

## 11. Fraud detection: flag transactions > 3× median


```python
transactions = np.array([100, 120, 5000, 110])
medianTxn = np.median(transactions)
fraudThreshold = 3 * medianTxn
fraudMask = transactions > fraudThreshold
flaggedTxns = transactions[fraudMask]
print(f"Median: {medianTxn} → Threshold: {fraudThreshold}")
print("Flagged transactions:", flaggedTxns)
```

## 12. Use np.ix\_ for multiple ROIs


```python
image100x100 = np.random.rand(100, 100)
roi1 = slice(20, 30), slice(30, 40)
roi2 = slice(60, 70), slice(70, 80)
rows = np.array([20, 60])
cols = np.array([30, 70])
rois = image100x100[np.ix_(rows, cols)]
print("Two 10×10 ROIs extracted → shape:", rois.shape)  # (2, 2) centers
```

## 13. Challenge: Extract checkerboard pattern


```python
checkerboard = np.random.randint(0, 2, size=(8, 8))
i, j = np.indices((8, 8))
blackSquares = checkerboard[(i + j) % 2 == 0]
print("Checkerboard 8×8:\n", checkerboard)
print(f"Black squares (i+j even): {len(blackSquares)} values")
```

## 14. Challenge: Modify original via fancy index


```python
arr = np.arange(10)
indices = [1, 3, 5]
arr[indices] = [99, 88, 77]
print("Modified array:", arr)
```

## 15. Challenge: Safe indexing with out-of-bounds


```python
arr = np.arange(5)
badIndex = [10]
safeIndex = np.clip(badIndex, 0, len(arr) - 1)
safeValue = arr[safeIndex]
print(f"Original: {arr}")
print(f"Tried index 10  clipped to {safeIndex[0]} value: {safeValue[0]}")
```
